```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('data.129.2.180.42.269.14.45.12.txt', sep=" ", header=None)
data = data.drop(0, axis=1).drop(3, axis=1)
data.columns = ['Date', 'Sigma']
```

```python
data.columns
```

```python
data['Sigma'].isna().value_counts()
```

```python
data[data['Sigma'].isna()]
df = data.loc[data['Sigma'].isna() == False]
```

```python
df
```

```python
x1 = df['Sigma'][20:]
```

```python
x2 = df['Sigma'][11:851]
```

```python
x3 = df['Sigma'][:840]
```

```python
x3.shape
```

```python
ax = plt.figure().add_subplot(projection='3d')

ax.plot(x1, x2, x3, lw=0.5)
ax.set_xlabel("X1 Axis")
ax.set_ylabel("X2 Axis")
ax.set_zlabel("X3 Axis")
ax.set_title("Climate Data")

plt.show()
```

```python
coords = np.vstack((x1, x2, x3))
print(coords.shape)
U, s, V = np.linalg.svd(coords)
U.shape, s.shape, V.shape
```

```python
print(np.vstack((x1, x2, x3)).shape)
basis = np.vstack((U[:,0], U[:,1], U[:,2]))
print(U)
print(basis)
coordssvd = np.linalg.solve(U, coords)
coordssvd
```

```python
ax = plt.figure().add_subplot(projection='3d')

ax.scatter(coordssvd[0], coordssvd[1], coordssvd[2], lw=0.5, alpha=.1)
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")

ax = plt.figure().add_subplot(projection='3d')

ax.plot(coordssvd[0], coordssvd[1], coordssvd[2], lw=0.5, alpha=.1)
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
```
