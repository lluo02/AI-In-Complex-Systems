```python
from tqdm.auto import tqdm
from scipy.spatial import cKDTree
data_2 = pd.DataFrame([coordssvd[0], coordssvd[1], coordssvd[2]]).transpose()
data_2 = data_2.to_numpy()
test_data = data_2[:600].copy()
```

```python
for t in tqdm(range(0, 172)):
    kdtree = cKDTree(data_2[:599 + t], leafsize=4)
    distances = np.array([0, 3, 3])
    indexes = np.array([0, 3, 3])
    for item in test_data:
        dists, idxs = kdtree.query(item, k=4, distance_upper_bound=.6)
        dists = dists[1:]
        idxs = idxs[1:]
        distances = np.vstack([distances, dists])
        indexes = np.vstack([indexes, idxs])

    distances = distances[1:]
    indexes = indexes[1:]
    
    predictions = [0, 0, 0]

    index = 599+t
    i1, i2, i3 = indexes[index, 0], indexes[index, 1], indexes[index, 2]
    d1 = data_2[i1+1] - data_2[i1] 
    d2 = data_2[i2+1] - data_2[i2]
    d3 = data_2[i3+1] - data_2[i3]
    d = (d1+d2+d3)/3
    
    predict = test_data[index] + d
    predictions = np.vstack([predictions, predict])

    predictions = predictions[1:]
    
    test_data = np.vstack([test_data, predictions])
```

```python
ax = plt.figure().add_subplot(projection='3d')
# ax.plot(data[:, 0], data[:, 1], data[:, 2], lw=.5, color='red')
ax.plot(test_data[600:, 0], test_data[600:, 1], test_data[600:, 2], lw=.5, color='blue')
plt.show()
```
